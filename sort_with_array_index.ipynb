# Sorting Algorithms Utilizing Array Indexes

```python
import numpy as np
```

## Simple Method

Use the array index directly as the value. Store a value at the index that represents the number of times that specific value occured in the input.

```python
# Use numpy to generate a random list of 10 integers between 0 and 100
inVals = np.random.randint(0, 100, 10)
print(inVals)

valueCount = np.ndarray(100, dtype=np.uint8)
for i in inVals:
    valueCount[i] += 1

sortedVals = []
for i in valueCount:
    for j in range(i):
        sortedVals.append(i)
     
valueCount
sortedVals
```
